```python
import requests
from binascii import hexlify, unhexlify

# Sending request to web server
payload = { 'plaintext' : 'secret message !' }
response = requests.post('http://3.90.169.27:8000/crypto2/', data=payload)

# Print response
ciphertext = response.text

for i in range(0, 256):
    c = ciphertext[:-2]
    c += format(i, 'x').zfill(2)

    # Sending request to web server
    payload = { 'ciphertext' : c }
    response = requests.post('http://3.90.169.27:8000/crypto2/', data=payload)

    # Print response
    print(c, response.text)
```
